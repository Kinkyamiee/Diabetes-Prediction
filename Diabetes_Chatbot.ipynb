```python
import torch
from transformers import pipeline

model_id = "meta-llama/Llama-3.2-1B"

pipe = pipeline(
    "text-generation",
    model=model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    token="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",
)
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-3.2-1B")
model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-3.2-1B",
                                             device_map="auto")
tokenizer.pad_token = tokenizer.eos_token
```

```python
generator = pipeline("text-generation",
                     model=model,
                     tokenizer=tokenizer,
                     max_new_tokens=100
                     )

def get_response(prompt):
  response = generator(prompt)
  gen_text = response[0]["generated_text"]
  return gen_text
```

```python
prompt = "What is diabetes?"
llama_response = get_response(prompt)
print(llama_response)
```

```python
import gdown
import os
import pickle
import streamlit as st
from datetime import date
from sklearn.preprocessing import StandardScaler

# Define Google Drive file ID and output path for the model
model_file_id = "1awxWYbRZ3M9WZFpDdXeuPUBx4SGWUTAB"  # Extracted file ID
model_path = "best_xgb_model.pkl"

# Download the model file if it doesn't exist locally
if not os.path.exists(model_path):
    gdown.download(f"https://drive.google.com/uc?id={model_file_id}", model_path, quiet=False)

# Load the trained model
with open(model_path, 'rb') as file:
    classifier = pickle.load(file)

# Path for scaler
scaler_file_id = "your_scaler_file_id_here"  # Replace with your scaler file ID
scaler_path = "scaler.pkl"

# # Download the scaler file if it doesn't exist locally
# if not os.path.exists(scaler_path):
#     gdown.download(f"https://drive.google.com/uc?id={scaler_file_id}", scaler_path, quiet=False)

# try:
#     with open(scaler_path, 'rb') as file:
#         scaler = pickle.load(file)
#     scale_features = True
# except FileNotFoundError:
#     st.error("Scaler file not found! Ensure the scaler is saved in the correct path.")
#     scale_features = False

# # Your Streamlit app code continues here...
# import os
# import pickle
# import streamlit as st
# from datetime import date
# from sklearn.preprocessing import StandardScaler
# from transformers import AutoTokenizer, AutoModelForCausalLM
# import torch

# # Load the XGBoost trained model
# model_path = r"C:\Users\hp\Documents\Diabetes_Health_Indicator\best_xgb_model.pkl"
# with open(model_path, 'rb') as file:
#     classifier = pickle.load(file)

# # Load the scaler used during model training
# scaler_path = r"C:\Users\hp\Documents\Diabetes_Health_Indicator\scaler.pkl"
# try:
#     with open(scaler_path, 'rb') as file:
#         scaler = pickle.load(file)
#     scale_features = True
# except FileNotFoundError:
#     st.error("Scaler file not found! Ensure the scaler is saved in the correct path.")
#     scale_features = False

# Define mappings for categorical inputs
sex_mapping = {"Male": 0, "Female": 1}
cholesterol_mapping = {"Yes": 1, "No": 0}
health_status_mapping = {
    "Excellent": 1,
    "Very good": 2,
    "Good": 3,
    "Fair": 4,
    "Poor": 5
}

# Function to calculate age from date of birth
def calculate_age(dob):
    today = date.today()
    return today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))

# Function to calculate BMI from feet and inches
def calculate_bmi(weight, feet, inches):
    total_height_meters = ((feet * 12) + inches) * 0.0254  # Convert height to meters
    if total_height_meters > 0:  # Avoid division by zero
        return round(weight / (total_height_meters ** 2), 2)
    return 0

# Prediction function for the diabetes model
@st.cache_data()
def prediction(bmi, age, phys_health, gen_health, ment_health, sex, high_chol):
    # Map categorical inputs to numerical values
    sex_numeric = sex_mapping[sex]
    high_chol_numeric = cholesterol_mapping[high_chol]
    gen_health_numeric = health_status_mapping[gen_health]

    # Prepare input data in the correct feature order
    input_data = [[bmi, age, phys_health, gen_health_numeric, ment_health, sex_numeric, high_chol_numeric]]

    # Apply feature scaling
    if scale_features:
        input_data_scaled = scaler.transform(input_data)
        prediction = classifier.predict(input_data_scaled)
    else:
        prediction = classifier.predict(input_data)

    if prediction == 0:
        pred = "You are not at risk of having diabetes."
    else:
        pred = "You are at risk of having diabetes."

    return pred

# LLaMA Model Integration
@st.cache_resource()
def load_llama_model():
    token = "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"
    tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-3.2-1B")
    model = AutoModelForCausalLM.from_pretrained(
        "meta-llama/Llama-3.2-1B", device_map="auto", torch_dtype=torch.float16
    )
    return tokenizer, model

# Text generation using LLaMA
@st.cache_data()
def generate_text(prompt, max_length=50):
    inputs = tokenizer(prompt, return_tensors="pt").to(llama_model.device)
    outputs = llama_model.generate(
        inputs.input_ids, max_length=max_length, num_return_sequences=1, temperature=0.7
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Main function for Streamlit app
def main():
    # Front end design
    html_temp = '''
    <div style='background-color: teal; padding:13px'>
    <h1 style='color: white; text-align: center;'>Diabetes Risk Prediction App</h1>
    </div>
    '''
    st.markdown(html_temp, unsafe_allow_html=True)

    # Section 1: Diabetes Risk Prediction
    st.subheader("Diabetes Risk Prediction")
    weight = st.number_input("Weight (kg)", min_value=0.0, step=0.1)
    feet = st.number_input("Height (feet)", min_value=0, step=1)
    inches = st.number_input("Height (inches)", min_value=0, max_value=11, step=1)
    bmi = calculate_bmi(weight, feet, inches)
    st.write(f"Calculated BMI: {bmi} (WHO classification)")

    dob = st.date_input("Date of Birth", min_value=date(1900, 1, 1), max_value=date.today())
    age = calculate_age(dob)
    st.write(f"Calculated Age: {age} years")

    phys_health = st.number_input(
        "Physical Health (how often have you fallen sick in the past month)", min_value=0, max_value=30
    )
    gen_health = st.selectbox("Rate your overall health condition", tuple(health_status_mapping.keys()))
    ment_health = st.number_input("Mental Health (days affected in the past month)", min_value=0, max_value=30)
    sex = st.selectbox("Sex", tuple(sex_mapping.keys()))
    high_chol = st.selectbox("High Cholesterol", tuple(cholesterol_mapping.keys()))

    if st.button("Predict Diabetes Risk"):
        result = prediction(bmi, age, phys_health, gen_health, ment_health, sex, high_chol)
        st.success(f"Prediction: {result}")

    # Section 2: LLaMA Text Generation
    st.subheader("LLaMA 3.2-1B Text Generation")
    tokenizer, llama_model = load_llama_model()
    user_prompt = st.text_input("Enter a prompt for LLaMA model:")
    if user_prompt:
        generated_text = generate_text(user_prompt)
        st.write("Generated Text:")
        st.write(generated_text)

if __name__ == '__main__':
    main()
```
